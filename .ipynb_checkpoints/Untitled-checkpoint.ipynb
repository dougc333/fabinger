```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
y=np.arange(0,10,.1)
print(y)
print(y.shape)
```

```python
z=np.random.rand(y.shape[0])
print(z)
print("-------")
slope=.2
new_z = slope*z
print(new_z)
```

```python
sm.OLS(new_z, y) 
```
